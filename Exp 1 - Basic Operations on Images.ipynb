# Experiment 1: Basic Operations on Images

```python
# Importing the Libraries
from skimage.io import imread
import cv2
import matplotlib.pyplot as plt
```

## 1. Access pixel values and modify them

```python
# Loading the Image
image = imread('Images/road.bmp', as_gray=True)
print('First 10 pixels from first row:', end=' ')
print(image[0][0:10])
plt.title('Original Image'), 
plt.imshow(image, cmap='gray') 
plt.show()
```

```python
# Modifying pixel values to un-normalize image pixels
normalized_image = image * 255
print('First 10 pixels from first row:', end=' ')
print(normalized_image[0][0:10])
plt.title('Normalized Pixel Values')
plt.imshow(normalized_image, cmap='gray')
plt.show()
```

## 2. Access image properties

```python
# Loading the Image & Accessing Properties
image = imread('Images/peppers.png')
print('Image shape:', image.shape)
print('Total number of pixels:', image.size)
plt.title('Image')
plt.imshow(image)
plt.show()
```

## 3. Image Region of Interest (ROI)

```python
# Loading the Image & Extracting ROI
image = imread('Images/peppers.png')
capsicum = image[180:480, 180:420]
plt.title('ROI: Capsicum')
plt.imshow(capsicum)
plt.show()
```

## 4. Splitting & Merging Image Channels

```python
# Loading Image & Splitting Image Channels
image = cv2.imread('Images/peppers.png')
b, g, r = cv2.split(image)
rgb_image = cv2.merge((r, g, b))
```

```python
# Displaying Different Image Channels
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax1.imshow(image)
ax1.title.set_text('BGR Channeled Image')
ax2.imshow(rgb_image)
ax2.title.set_text('RGB Channeled Image')
plt.show()
```
