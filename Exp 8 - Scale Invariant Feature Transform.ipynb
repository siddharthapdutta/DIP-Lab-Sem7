# Experiment 8: Scale-Invariant Feature Transform (SIFT)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading Input Image
image = cv2.imread('Images/home.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Detecting Keypoints
siftObj = cv2.xfeatures2d.SIFT_create()
keyPoints, descriptors = siftObj.detectAndCompute(image, None)
```

```python
# Drawing Keypoints on Input Image
result = cv2.drawKeypoints(image, keyPoints, None)
```

```python
# Displaying Results
fig = plt.figure(figsize=(15, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.imshow(image, cmap='gray')
ax1.title.set_text('Original Image')
ax2.imshow(result)
ax2.title.set_text('Image with Keypoints using SIFT')
```
