# Experiment 2: Making Borders for Images (Padding)

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading Source Image
src = cv2.imread('Images/road.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# Adding Different Borders
# cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value)
replicated = cv2.copyMakeBorder(src, 20, 20, 20, 20, cv2.BORDER_REPLICATE)
reflected = cv2.copyMakeBorder(src, 20, 20, 20, 20, cv2.BORDER_REFLECT)
reflected101 = cv2.copyMakeBorder(src, 20, 20, 20, 20, cv2.BORDER_REFLECT_101)
wrapped = cv2.copyMakeBorder(src, 20, 20, 20, 20, cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(src, 20, 20, 20, 20, cv2.BORDER_CONSTANT, value=[255, 255, 255]) # White Color Value
```

```python
# Displaying Results
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
ax6 = fig.add_subplot(236)

ax1.imshow(src, cmap='gray')
ax1.title.set_text('Original Image')
ax2.imshow(replicated, cmap='gray')
ax2.title.set_text('Replicated Border')
ax3.imshow(reflected, cmap='gray')
ax3.title.set_text('Reflected Border')
ax4.imshow(reflected101, cmap='gray')
ax4.title.set_text('Reflected 101 Border')
ax5.imshow(wrapped, cmap='gray')
ax5.title.set_text('Wrapped Border')
ax6.imshow(constant, cmap='gray')
ax6.title.set_text('Constant Border (White)')
plt.show()
```
